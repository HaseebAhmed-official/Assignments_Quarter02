## **6. Download the Image from the Internet üåê**

```python
url="https://images.unsplash.com/photo-1719937051230-8798ae2ebe86?q=80&w=2072&auto=format&fit=crop&ixlib=rb-4.0.3&ixid=M3wxMjA3fDF8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
response=requests.get(url)
img=PIL.Image.open(BytesIO(response.content))
```

## **11. Experiment with a Different AI Model and Configuration**

```python
from google.generativeai.generative_models import GenerativeModel
model: GenerativeModel = genai.GenerativeModel(
    'gemini-1.5-flash',
     generation_config=genai.GenerationConfig(
        max_output_tokens=200,
        temperature=1,
    ))
response=model.generate_content([prompt,img])
display(Markdown(response.text))
```

Switches to the **gemini-1.5-flash** model.

### **Adjusts the generation configuration:**
**max_output_tokens:** Limits the length of the AI's response.

**temperature:** Controls creativity (higher = more creative).
* Sends the prompt and image to the new model for processing  
* Displays the new response as markdown.

**Why it's important:** Tests a different model and settings to compare results and improve output quality.
